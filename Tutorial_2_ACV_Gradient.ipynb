```python
import numpy as np
import matplotlib.pyplot as plt
x,y=0.0,0.0
a=3
b=100
learning_rate=0.0001

def f(x,y):
    return (a - x)**2+b*(y - x**2)**2


new_x,new_y=0.0,0.0
threshold=1e-6
trajectory_x=[]
trajectory_y=[]
trajectory_f=[]

for i in range(2000000):
    dx=-2*(a - x)-4*x*b*(y - x**2)
    dy=2*b*(y-x**2)
    new_x=x-learning_rate*dx
    new_y=y-learning_rate*dy
    trajectory_x.append(new_x)
    trajectory_y.append(new_y)
    trajectory_f.append(f(new_x,new_y))

    #if abs(x-new_x) < threshold and abs(y-new_y) < threshold:
    if np.sqrt(dx**2 + dy**2) < 1e-6:
        break

    x,y=new_x,new_y

print("Final point:", new_x, new_y)
print("Minimum should be:", a, a**2)
print("Iterations:", len(trajectory_x))

plt.plot(trajectory_x,trajectory_y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient descent path")
plt.show()

plt.plot(trajectory_f)
plt.xlabel("iterations")
plt.ylabel("f(x,y)")
plt.title("Objective value over iterations")
plt.show()

plt.plot(trajectory_x, label="x")
plt.plot(trajectory_y, label="y")
plt.xlabel("Iteration")
plt.ylabel("Parameter value")
plt.title("x and y over iterations")
plt.legend()
plt.show()

```
